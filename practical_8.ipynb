```
import os
os.environ['TF_USE_LEGACY_KERAS']='1'
import tf_keras
#Importing required libraries
import tensorflow as tf
import torch
import nltk
import pandas as pd
import matplotlib.pyplot as plt

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from transformers import (
    BertTokenizer,
    BertModel,
    BertForMaskedLM,
    TFBertForSequenceClassification,
    pipeline
)
from bertviz import head_view
```

```
#Download nltk data
nltk.download('punkt')
nltk.download('stopwords')
```

```
#Stopword set
sw=set(stopwords.words('english'))
txt_arr=[
    "The coffee machine stopped working after two days",
    "I absolutely loved the soundtrack of this series",
    "The train reached exactly at the scheduled time",
    "The keyboard keys feel smooth and comfortable",
    "The soup tasted stale and unpleasant",
    "My internet speed has been stable all week",
    "The smartwatch display looks very sharp",
    "The lecture today was neither boring nor exciting",
    "The shoe sole came off within a month",
    "The park was quiet during the evening"
]
```

```
lb = [1,0,2,1,0,2,1,2,0,2]
```

```
tk=BertTokenizer.from_pretrained('bert-base-uncased')
```

```
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
clr_txt=[]
for sent in txt_arr:
    wrd=word_tokenize(sent.lower())
    tmp=[]
    for j in wrd:
        if j not in sw:
            tmp.append(j)
    clr_txt.append(tmp)
```

```
#Print cleaned text
print("After removing stopwords:\n")
for i in clr_txt:
    print(i)
```

```
print("BERT Tokens:")
for s in txt_arr:
    enc=tk.tokenize(s)
    print(enc)
```

```
x=tk(txt_arr,padding=True,truncation=True,return_tensors='tf')

print("Input IDs:")
print(x['input_ids'])
```

```
bert_m=BertModel.from_pretrained('bert-base-uncased',output_attentions=True)

for line in txt_arr[:2]:
    ip=tk(line, return_tensors='pt')
    op=bert_m(**ip)
    attn=op.attentions
    tok=tk.convert_ids_to_tokens(ip['input_ids'][0])
    head_view(attn, tok)
```

```
mk=pipeline("fill-mask",model="bert-base-uncased")
msk_sent="The movie was [MASK]"
res=mk(msk_sent)

print("Masked Word Predictions:")
for r in res:
    print(r['token_str'], ":", r['score'])
```

```
txt1="The camera quality is excellent"
ip2=tk(txt1, return_tensors='pt')
op2=bert_m(**ip2)
att=op2.attentions
tok2=tk.convert_ids_to_tokens(ip2['input_ids'][0])
ly=0
hd=0
mat=att[ly][0][hd].detach().numpy()
df=pd.DataFrame(mat, columns=tok2, index=tok2)

print("Attention Matrix Dataframe:")
print(df)
```

```
mdl=TFBertForSequenceClassification.from_pretrained('bert-base-uncased',num_labels=3)
```

```
opt=tf.keras.optimizers.Adam(learning_rate=3e-5)
```

```
mdl.compile(
    optimizer=opt
)
```

```
y=tf.constant(lb)

mdl.fit(
    {
        'input_ids': x['input_ids'],
        'attention_mask': x['attention_mask']
    },
    y,
    epochs=15
)
```

```
test=["The smartwatch display is amazing"]

test_ip=tk(test,padding=True,truncation=True,return_tensors='tf')

prd=mdl.predict(dict(test_ip))
cls=tf.argmax(prd.logits, axis=1).numpy()[0]

sent_map = {0: "Negative",1: "Positive",2: "Neutral"}

print("Predicted Sentiment :")
print(sent_map[cls])
```
